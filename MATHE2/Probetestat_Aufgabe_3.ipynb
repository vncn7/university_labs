```python
import numpy as np
import matplotlib.pyplot as plt

# Punkte definieren
points = np.array([[1, 0], [2, 0], [3, 1], [4, 0], [5, 0]])

# LGS-Koeffizientenmatrix erstellen
A = np.vander(points[:, 0], len(points))

# Rechte Seite des LGS erstellen
b = points[:, 1]

# LGS lösen
coefficients = np.linalg.solve(A, b)
print(coefficients)

# Polynomfunktion erstellen
poly_func = np.poly1d(coefficients)

# x-Werte für die Kurve generieren
x = np.linspace(0, 6, 100)

# y-Werte für die Kurve berechnen
y = poly_func(x)

# Plot erstellen
plt.scatter(points[:, 0], points[:, 1], color='red', marker='*')
plt.plot(x, y, color='blue')

# Achsenbeschriftungen
plt.xlabel('x')
plt.ylabel('y')

# Achsenbereiche festlegen
plt.xlim(0, 6.5)
plt.ylim(-1, 2.5)

# Achsen-Schrittweiten festlegen
plt.xticks(np.arange(0, 7, 1))
plt.yticks(np.arange(-1, 3, 1))

# Seitenverhältnis beibehalten
plt.gca().set_aspect('equal')

# Gitter anzeigen
plt.grid(True)

# Diagramm anzeigen
plt.show()
```
