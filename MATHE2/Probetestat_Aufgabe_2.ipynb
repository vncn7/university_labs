```python
import sympy as sp

# Symbole definieren
t = sp.symbols('t')
x = sp.Function('x')(t)

# Ableitungen
dx = sp.Derivative(x)
ddx = sp.Derivative(dx)

# DGL x''(t) + 3x'(t) + 2x(t) = 2sin(t)
ode = sp.Eq(ddx+3*dx+2*x, 2*sp.sin(t))
print("DGL mit den Anfangswerten: x(0) = 0, x(0) = 1")
display(ode)

# Allgemeine Lösung der homogenen Differenzialgleichung:
print("\nAllgemeine Lösung der homogenen Differenzialgleichung:")
display(sp.dsolve(sp.Eq(ode.lhs, 0)))

# Allgemeine Lösung der inhomogenen Differenzialgleichung:
print("\nAllgemeine Lösung der inhomogenen Differenzialgleichung:")
display(sp.dsolve(ode))

# Lösung des Anfangswertproblems x(0)=0, x'(0)=1:
initconds = {x.subs(t, 0): 0, dx.subs(t, 0): 1}
print("\nLösung des Anfangswertproblems:")
sol = sp.dsolve(ode,ics=initconds)
display(sol)

# Skizze des Anfangswertproblems:
print("Skizze der Lösung des Anfangswertproblems:")
sp.plot(sol.rhs, (t, 0, 10))
```
