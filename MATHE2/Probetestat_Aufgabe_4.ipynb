```python
import numpy as np
import matplotlib.pyplot as plt

# Definiere die minimalen und maximalen Werte für x- und y-Koordinaten
x_min, x_max, y_min, y_max = -4.0, 4.5, -5.0, 7.5

# Erstelle ein Figure- und Axes-Objekt mit einer bestimmten Größe
fig, ax = plt.subplots(figsize=((x_max-x_min)/2, (y_max-y_min)/2))

# Setze das Seitenverhältnis des Plots auf gleich
ax.set_aspect('equal')

# Setze die Begrenzungen der x- und y-Achsen
ax.axis([x_min, x_max, y_min, y_max])

# Erzeuge Markierungen für die x-Achse
x_ticks = np.arange(x_min, x_max)
ax.set_xticks(x_ticks)

# Erzeuge Markierungen für die y-Achse
y_ticks = np.arange(y_min, y_max)
ax.set_yticks(y_ticks)

# Zeige Gitterlinien im Plot an
ax.grid(True)

# Plotte drei Linien im Diagramm
plt.plot([-3, -3], [-3, 2], color='black')
plt.plot([3, 3], [2, -3], color='black')
plt.plot([-3, 3], [2, 2], color='black')
plt.plot([-3, 3], [-3, -3], color='black')
plt.plot([0, -3], [6, 2], color='black')
plt.plot([0, 3], [6, 2], color='black')

# Erzeuge Arrays für die x- und y-Koordinaten der Linien
x_1 = np.array([-3, -3])
y_1 = np.array([-3, 2])
x_2 = np.array([3, 3])
y_2 = np.array([-3, 2])
x_3 = np.array([-3, 3])
y_3 = np.array([-3, -3])
x_4 = np.array([-3, 3])
y_4 = np.array([2, 2])
x_5 = np.array([-3, 0])
y_5 = np.array([2, 6])
x_6 = np.array([3, 0])
y_6 = np.array([2, 6])

# Erzeuge ein Array mit NaN (Not a Number)-Werten für Lücken zwischen den Linien
nan = np.array([np.nan])

# Verkette die x- und y-Koordinaten mit NaN-Werten
X = np.concatenate((x_1, nan, x_2, nan, x_3, nan, x_4, nan, x_5, nan, x_6))
Y = np.concatenate((y_1, nan, y_2, nan, y_3, nan, y_4, nan, y_5, nan, y_6))
XY = np.array([X, Y])

# Gib die kombinierten Koordinaten aus
print(XY)

# Konvertiere Grad in Bogenmaß für die Rotation
Umrechnungsfaktor = 2 * np.pi / 360
alpha_in_grad = 20
alpha_in_bog = alpha_in_grad * Umrechnungsfaktor

# Erzeuge eine Rotationsmatrix
R = np.array([[np.cos(alpha_in_bog), -np.sin(alpha_in_bog)],
              [np.sin(alpha_in_bog), np.cos(alpha_in_bog)]])

# Wende die Rotationsmatrix auf die kombinierten Koordinaten an
XY_rot = R @ XY

# Plotte die rotierten Koordinaten
ax.plot(XY_rot[0, :], XY_rot[1, :], 'r')

# Zeige den Plot an
plt.show()
```
